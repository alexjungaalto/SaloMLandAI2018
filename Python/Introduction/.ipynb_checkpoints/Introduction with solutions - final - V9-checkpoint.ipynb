```python
### SOLUTION ###

import numpy as np # define shorthand "np" for the numpy package 
import matplotlib.pyplot as plt #define shorthand "plt" for package matlobplit.pyplot

def example_func(x):
    f_x = 1/(1+np.exp(-x))
    return f_x

range_x = np.arange(-6 , 6 , 0.01)

f_x = np.empty(len(range_x))

for i in range(len(range_x)):
    f_x[i] = example_func(range_x[i])

### STUDENT TASK ###
#1) Solution:
plt.plot(range_x,f_x, label='sigmoid function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

### STUDENT TASK ###
#2) Solution:
f_prime_x = f_x * (1-f_x)
plt.plot(range_x,f_prime_x, label='derivative of sigmoid function')
plt.xlabel('x')
plt.ylabel('f(x)(1-f(x))')
plt.legend()
plt.show()
```

## Question 3. Eigendecomposition

Consider a dataset containing $N$ feature vectors $\mathbf{x}^{(1)},...,\mathbf{x}^{(N)} \in \mathbb{R}^{d}$ which we stack into the feature matrix $\mathbf{X}=(\mathbf{x}^{(1)},\ldots,\mathbf{x}^{(N)})^{T} \in \mathbb{R}^{N \times d}$. In order to characterize the shape of dataset it is useful to compute the sample covariance matrix $$\mathbf{C} = (1/N) \mathbf{X}^{T} \mathbf{X}.$$ 
The so obtained matrix $\mathbf{C}$ is positive semidefinite (psd) since $\mathbf{w}^{T}\mathbf{C}\mathbf{w} \geq0 $ for any vector $\mathbf{w} \in \mathbb{R}^{d}$. As a psd matrix, we can decompose $\mathbf{C}$ into three factors 
$$\mathbf{C} = \mathbf{U} {\bf \mathbf{\Lambda}} \mathbf{U}^{T}.$$ 
The matrix $\mathbf{U}$ is orthogonal (i.e., $\mathbf{U}^{T}\mathbf{U}=\mathbf{I}$) and its columns are eigenvectors of $\mathbf{C}$. The matrix  ${\mathbf{\Lambda}}$ is diagonal having the eigenvalues of $\mathbf{C}$ on the main diagonal.

```python
### SOLUTION ###

import numpy as np
import matplotlib.pyplot as plt

# Set the random seed so that the data is always the same
np.random.seed(1234) #Do not change the seed number!
#generate the data
X = np.random.multivariate_normal([0,0], [[1, 0.5], [0.5, 0.5]], 1000)

### STUDENT TASK ###
#Step 1: draw the scatter plot for data X
plt.figure(figsize=(8,8))
plt.scatter(X[:,0], X[:,1], label='data')


### STUDENT TASK ###
#Step 2: compute the sample covariance of data X
C = np.matmul(np.transpose(X), X)/X.shape[0]

### STUDENT TASK ###
#Step 3: find eigenvalues and eigenvectors of C and print them
values, vectors = np.linalg.eig(C)
print('eigenvalues= ', values)
print('eigenvectors= ', vectors)

### STUDENT TASK ###
#Step 4: plot each eigenvector as a line from [0,0] to the coordinate of the eigenvevtor
plt.plot([0,vectors[0,1]],[0,vectors[0,0]],'r-', label='eigenvectors')
plt.plot([0,vectors[1,1]],[0,vectors[1,0]],'r-')

plt.xlim((-4, 4))
plt.ylim((-4, 4))

plt.legend()
plt.show()
```
